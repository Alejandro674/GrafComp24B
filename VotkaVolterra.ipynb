# MODELO VOTKA VOLTERRA
## ¿QUE ES EL MODELO?

 El modelo VOTKA VOLTERRA es un modelo de ecuaciones diferenciales que define el cómo funciona la presa y el depredador
Las ecuaciones de Lotka-Volterra son un par de ecuaciones diferenciales de primer orden, no lineales, que se utilizan para describir la dinámica de sistemas biológicos en los que interactúan dos especies: una depredadora y otra presa. Estas ecuaciones capturan la interacción entre ambas poblaciones, modelando su crecimiento y declive en función de las tasas de depredación y reproducción.

```python
import matplotlib.pyplot as plt
from random import *
from numpy import *
import sys
```

```python
a = 0.7; b = 0.3; c = 0.3;  e = 0.9 #ejercicio 1
dt = 0.001; max_time = 100

# initial time and populations
t = 0; x = 1.0; y = 1.0
```

```python
# empty lists in which to store time and populations
t_list = []; x_list = []; y_list = []

# initialize lists
t_list.append(t); x_list.append(x); y_list.append(y)

while t < max_time:
    # calc new values for t, x, y
    t = t + dt
    x = x + (a*x - b*x*y)*dt
    y = y + (-c*y + e*x*y)*dt

    # store new values in lists
    t_list.append(t)
    x_list.append(x)
    y_list.append(y)

# Plot the results
p = plt.plot(t_list, x_list, 'g', t_list, y_list, 'r', linewidth = 2)
```

Como se observa, cuando incrementan las presas incrementa un
poco los depredadores, estos depredadores crecen significativamente, y observamos como necesitamos de más variables para que sea constante, y una mejor predicción del mundo  real, dado que no es posible que mientras existan menos presas, aumente el número de depredadores
