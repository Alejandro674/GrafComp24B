Ahora toca hacer las magnitudes de los distintos angulos por ejemplo 

si tenemos B = (-5,2)

entonces si tenemos b^2 = c^2+a^2

por lo que tenemos que despegar b = 5.4 aproximadamente

```python
x1 = -10
x2 = 10

y1= -10
y2 = 10

plt.axis([x1,x2,y1,y2])
plt.axis("on")
plt.grid(True)
plt.title("Plano Cartesiano")
plt.text(-3.75,-0.55, "||B||",color="r")
plt.text(-2.5,-3, "Bx",color="k",fontstyle = "italic" )
plt.text(0.75,-1.5, "By",color="k",fontstyle = "italic" )

plt.arrow(0,0,-4.85,-2,head_length=0.25,head_width=0.25,color="r")
plt.arrow(-5,-2,5,0,head_length=0,head_width=0,color="k",linestyle=":")
plt.arrow(-0.1,-2,0,2,head_length=0,head_width=0,color="k",linestyle = ":")

```

vemos como el vector B es rojo y obtenemos los Bx y By de los distintos traslados de la magnitud

```python
x1 = -10
x2 = 10

y1= -10
y2 = 10

plt.axis([x1,x2,y1,y2])
plt.axis("on")
plt.grid(True)
plt.title("Plano Del vector C")
plt.text(1.5,-2, "||C||",color="r")
plt.text(-1,-2, "Cx",color="k",fontstyle = "italic" )
plt.text(0.5,-5, "Cx",color="k",fontstyle = "italic" )

plt.arrow(0,0,2,-4,head_length=0.25,head_width=0.25,color="r")
plt.arrow(0,-4,2,0,head_length=0,head_width=0,color="k",linestyle=":")
plt.arrow(0.05,-4,0,4,head_length=0,head_width=0,color="k",linestyle = ":")

```

Observamos como se trasladan tanto en X como en Y por medio de Cx y Cy para su posible graficacion 

donde el vector C es (2,-4)

si hacemos la ecuacion quedaria raiz cuadrada de 20 entonces la magnitud del vector es sqrt(20) o raiz de 20

```python
x1 = -10
x2 = 10

y1= -10
y2 = 10

plt.axis([x1,x2,y1,y2])
plt.axis("on")
plt.grid(True)
plt.title("Plano Cartesiano")
plt.text(-3.75,-0.55, "||D||",color="r")
plt.text(-3.5,-4, "Dx",color="k",fontstyle = "italic" )
plt.text(0.75,-1.5, "Dy",color="k",fontstyle = "italic" )

plt.arrow(0,0,-7,-3,head_length=0.25,head_width=0.25,color="r")
plt.arrow(-7,-3,7,0,head_length=0,head_width=0,color="k",linestyle=":")
plt.arrow(-0.05,-3,0,3,head_length=0,head_width=0,color="k",linestyle = ":")

```

ahora observamos como quedaria el ultimo vector, D = (-7,-3)

Donde su magnitud es de raiz de 58 

y por ultimo su diagonal es para los numeros negativos
