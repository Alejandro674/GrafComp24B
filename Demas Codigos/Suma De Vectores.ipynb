```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

Importamos las librerias necesarias, en este caso, random, numpy y matplotlib


```python
x1 = -10
x2 = 10

y1= -10
y2 = 10

plt.axis([x1,x2,y1,y2])
plt.axis("on")
plt.grid(True)
plt.title("Plano Cartesiano de Vectores")
plt.text(2,2, "||A||",color="r")
plt.text(-1.5,4, "||B||",color="b")
plt.text(-4.5,-1.5, "||C||",color="g")
plt.text(0,-3, "||A||+||B||+||C||")
plt.arrow(0,0,2,4,head_length=0.25,head_width=0.25,color="r")
plt.arrow(2,4,-6,-2,head_length=0.25,head_width=0.25,color="b")
plt.arrow(-4,2,2,-7,head_length=0.25,head_width=0.25,color="g")
plt.arrow(0,0,-2,-5,head_length=0.25,head_width=0.25)



```

realizamos el cuadro cartesiano, desde -10 hasta 10, despues obtenemos los nombres de los vectores llamados ||A||, ||b||, ||c||, ||d||, posteriormente le damos la direccion y posicionamiento de los vectores, con variaciones, para posteriormente hacer una graficacion del vector final, que en este caso es la suma,
